```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statsmodels.tsa.api import Holt
# set all the data to 4 decimal
pd.set_option("display.float.format", lambda x: "%.2f" % x) 
```

```python
# define function to select relevant dataframe
def display_relevant(df):
    rowname = ["United States", "China", "Germany", "Japan","India",
               "United Kingdom", "France", "Italy", "Brazil","Canada"]
    colname = ["2011", "2012", "2013", "2014", "2015", "2016"]
    df = df.loc[rowname, colname]
    # change the data type to float
    #df = df.iloc[0:11, 0:7].astype(dtype="float64")
    return df
```

# Growth healthcare

```python
def growth(df):
    df["growth"] = df.iloc[:,-1] - df.iloc[:,0]
    df["growth in percentage"] = df["growth"]/df.iloc[:,0]
    return df
```

```python
growth_health = growth(df_health)
growth_health
```

# Growth education

```python
growth_education = growth(df_edu)
growth_education
```

# Predict total

```python
def total_predict(df):
    rowname = ["United States", "China", "Germany", "Japan","India",
               "United Kingdom", "France", "Italy", "Brazil","Canada"]
    df = df.loc[rowname,:]
    df = df.T.loc["2000":"2016"]
    df = df.T
    data = list(df.sum())
    index= pd.date_range(start='2000', end='2017', freq='A')
    df_clean = pd.Series(data, index)
    return df_clean
```

```python
edu_predict = total_predict(df_4)/1000000000
edu_predict
```

```python
fit = Holt(edu_predict, damped=True).fit(smoothing_level=0.8, smoothing_slope=0.2)
fcast = fit.forecast(5).rename("Holt's Additive damped trend")

ax = edu_predict.plot(color="black", marker="o", figsize=(12,8))
fit.fittedvalues.plot(ax=ax, color='green')
fcast.plot(ax=ax, color='green', marker="o", legend=True)

plt.show()
```

```python
fcast
```

# Predict Country

```python
def predict(df, country):
    df = df.T[country]
    df = pd.DataFrame(df.loc["2000":"2016"])
    data = list(df[country])
    index= pd.date_range(start='2000', end='2017', freq='A')
    df_clean = pd.Series(data, index)
    return df_clean
```

```python
gdp_usa = predict(df_1, "United States")
gdp_china = predict(df_1,"China")
edu_usa = predict(df_4, "United States")
edu_china = predict(df_4,"China")
health_usa = predict(df_3,"United States")*gdp_usa/100/1000000000
health_china = predict(df_3, "China")*gdp_china/100/1000000000 
military_usa = predict(df_2,"United States")*gdp_usa/100/1000000000
military_china = predict(df_2, "China")*gdp_china/100/1000000000 
```

```python
fit = Holt(edu_china, damped=True).fit(smoothing_level=0.8, smoothing_slope=0.2)
fcast = fit.forecast(5).rename("Holt's Additive damped trend")

ax = edu_china.plot(color="black", marker="o", figsize=(12,8))
fit.fittedvalues.plot(ax=ax, color='green')
fcast.plot(ax=ax, color='green', marker="o", legend=True)

plt.show()
```
